```python
import requests
from bs4 import BeautifulSoup


# Making a GET request
r = requests.get('https://www.nature.com/articles/cdd201489')

# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')

s = soup.find('div', class_='entry-content')
content = s.find('p')
 
print(content)
```
